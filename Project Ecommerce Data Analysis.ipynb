## **Database Connectivity:**

To perform the analysis, we'll establish a connection to the MySQL database containing
the ecommerce data. We'll use the appropriate Python libraries, such as pandas and
mysql, to fetch and manipulate the data directly from the database.

```python
import mysql.connector
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def mydbconnect():
     mydb = mysql.connector.connect (
     host="localhost",
     user="root",
     password="ROOT",
     database="ecommerce"
      )
     return mydb
```

```python
mydb=mydbconnect()
mycursor = mydb.cursor()

mycursor.execute("show tables")

myresult = mycursor.fetchall()

for x in myresult:
  print(x)

mydb.close()
```

```python
# Create a cursor object to execute SQL queries
mydb=mydbconnect()
mycursor = mydb.cursor()
mycursor.execute("select * from customer")
customer_data = pd.DataFrame(mycursor.fetchall(), columns=[desc[0] for desc in
mycursor.description])
# Query data from the 'product' table
mycursor.execute('SELECT * FROM product')
product_data = pd.DataFrame(mycursor.fetchall(), columns=[desc[0] for desc in
mycursor.description])
# Query data from the 'order_details' table
mycursor.execute('SELECT * FROM order_details')
order_data = pd.DataFrame(mycursor.fetchall(), columns=[desc[0] for desc in
mycursor.description])
#printing first 5 records from each table
print(customer_data.head())
print(" ")
print(product_data.head())
print(" ")
print(order_data.head())
```

## **Data Cleaning:**

Before proceeding with the analysis, let's perform some basic data cleaning:

**Removing Missing Values**

```python
print("Before Removing nan Values from customer table")
print(customer_data.info())
#removing the nan values from customer table
customer_data.dropna(inplace=True)
print("After Removing nan Values from customer table ")
print(customer_data.info())
print("Before Removing nan Values from product table")
print(product_data.info())
#removing the nan values from product table
product_data.dropna(inplace=True)
print("After Removing nan Values from product table")
print(product_data.info())
print("Before Removing nan Values from order table")
print(order_data.info())
#removing the nan values from order table
order_data.dropna(inplace=True)
print("After Removing nan Values from order table")
print(order_data.info())
```

**Removing Duplicates**

```python
#Removing duplicate values
customer_data.drop_duplicates(inplace=True)
product_data.drop_duplicates(inplace=True)
order_data.drop_duplicates(inplace=True)
#Find the duplicate values
duplicate_rows1 = customer_data.duplicated()
duplicate_rows2 = product_data.duplicated()
duplicate_rows3 = order_data.duplicated()
#Printing the no of duplicate rows
print("After Removing Number of duplicate rows from customer table:", duplicate_rows1.sum())
print(customer_data[duplicate_rows1])
print("After Removing Number of duplicate rows from product table:", duplicate_rows2.sum())
print(product_data[duplicate_rows2])
print("After Removing Number of duplicate rows from order table:", duplicate_rows3.sum())
print(order_data[duplicate_rows3])
```

## **Exploratory Data Analysis (EDA) and Visualization:**


**● Customer Analysis:**

* Identify the total number of customers City wise.

```python
colors = ['red', 'orange', 'brown', 'pink','yellow','green','violet','blue','purple','seagreen','skyblue']
customer_data.groupby('city')['customer_id'].count().sort_values(ascending=False).plot(kind='bar',xlabel='City',ylabel='Number of customers',title='Number of Customers city-wise',color=colors)
```

* Identify the most frequent customers based on their order history

```python
order_data.groupby('customer_id')['order_id'].count().sort_values(ascending=False).plot(kind='bar',xlabel='customer_id',ylabel='Number of customers',title='Most frequent customers')
```

**● Product Analysis:**
*  Determine the total number of products available by category

```python
product_data.groupby('category')['product_id'].count().sort_values(ascending=False).plot(kind='bar',xlabel='Category',ylabel='Number of occurrences',title='Distribution of Products',color=['orange','yellow','green','violet'])
```

* Analyze the distribution of products across sub-categories.

```python
product_data.groupby('sub_category')['product_id'].count().sort_values(ascending=False).plot(kind='bar',xlabel='Sub-Category',ylabel='Number of Products',title='Distribution of Products Products Across Sub-Categories',color='pink')
```

* Identify products with low stock levels

```python
product_data.groupby('product_name')['stock'].sum().sort_values(ascending=True).head(5).plot(kind='bar',xlabel='Product Name',ylabel='Stock Levels',title='Product with Low Stock Levels',color='pink')
```

* Calculate the average, maximum, and minimum selling prices for products.

```python
ave=product_data['selling_price'].mean()
max=product_data['selling_price'].max()
min=product_data['selling_price'].min()
mathematical_operators=['Average','Maximum','Minimum']
values=[ave,max,min]
plt.bar(mathematical_operators,values,color=['orange','purple','seagreen'])
# Add labels and a title
plt.ylabel("Price")
plt.title("Average, Maximum,Minimum Price for Product")
# Display the plot
plt.show()
```

**● Order Analysis:**
* Calculate the top 10 orders product wise.

```python
mydb=mydbconnect()
mycursor = mydb.cursor()
mycursor.execute("select p.product_name,p.category,o.quantity,o.total_price,o.order_id from product p natural join order_details o")
product_Wise_order_data = pd.DataFrame(mycursor.fetchall(), columns=[desc[0] for desc in
mycursor.description])
product_Wise_order_data.groupby('product_name')['order_id'].sum().sort_values(ascending=False).head(10).plot(kind='bar',xlabel='Product Name',ylabel='Number of Orders',title='Total Number of Orders Product_Wise with Product Name',color='skyblue')
```

* Analyze the order status distribution (e.g., pending, delivered)

```python
order_data.groupby('order_status')['order_id'].count().sort_values(ascending=False).plot(kind='bar',xlabel='Orfer Status',ylabel='Number of Orders',title='Order Status Distribution',color=['green','orange','red'])
```

* Identify the most popular products based on order quantity.

```python
product_Wise_order_data.groupby('product_name')['quantity'].sum().sort_values(ascending=True).head(10).plot(kind='barh',xlabel='Total Order Quantity',ylabel='Product Name',title='Top 10 Most Popular Products based on Order Quatity',color='skyblue')
```

 ● **Sales Analysis:**
* Calculate total revenue generated from orders product wise

```python
product_Wise_order_data.groupby('product_name')['total_price'].sum().sort_values(ascending=False).head(10).plot(kind='bar',xlabel='Product Name',ylabel='Total Revenue',title='Top 10 Products by Total Revenue',color='magenta')
```

* Calculate the total revenue generated from all orders

```python
Total_revenue=order_data['total_price'].sum()
plt.bar('Total Revenue Metrics',Total_revenue)
# Add labels and a title
plt.ylabel("Total Revenue")
plt.title("Total Revenue Generated from all Orders")
# Display the plot
plt.show()
```

*  Identify the most profitable products based on the difference between

```python
product_data.groupby('product_name')['original_price'].sum().sort_values(ascending=False).head(10).plot(kind='bar',xlabel='Product Name',ylabel='Profit(Selling Price-Orginal Price)',title='Top Most Profitable Products',color='seagreen')
```

● **Customer Order Patterns:**
* Identify product names with the highest and lowest order quantities.

```python
df_sorted_descending = product_Wise_order_data.sort_values(by='quantity', ascending=False)

# Extract the top and bottom N records (you can choose N based on your preference)
top_n=1
bottom_n=6
top_products = df_sorted_descending.head(top_n)
bottom_products = df_sorted_descending.tail(bottom_n)

# Combine the top and bottom DataFrames
combined_df = pd.concat([top_products, bottom_products])

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(combined_df['product_name'], combined_df['quantity'], color=['green']*top_n + ['red']*bottom_n)
plt.xlabel('Product Name')
plt.ylabel('Total Quantity Ordered')
plt.title('Total Quantity of Highest and Lowest Product')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

* Identify customers with the highest and lowest order quantities by
customer name.

```python
mydb=mydbconnect()
mycursor = mydb.cursor()
mycursor.execute(" select c.name,o.quantity from customer c natural join order_details o")
customer_Wise_order_data = pd.DataFrame(mycursor.fetchall(), columns=[desc[0] for desc in
mycursor.description])

df_sorted_descending = customer_Wise_order_data.sort_values(by='quantity', ascending=False)

# Extract the top and bottom N records (you can choose N based on your preference)
top_n=1
bottom_n=6
top_products = df_sorted_descending.head(top_n)
bottom_products = df_sorted_descending.tail(bottom_n)

# Combine the top and bottom DataFrames
combined_df = pd.concat([top_products, bottom_products])

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(combined_df['name'], combined_df['quantity'], color=['green']*top_n + ['red']*bottom_n)
plt.xlabel('Customer Name')
plt.ylabel('Total Quantity Quantity')
plt.title('Total Order Quantity by Customer(Highest and Lowest)')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

* Determine the most preferred payment modes

```python
order_data.groupby('payment_mode')['order_id'].count().sort_values(ascending=False).plot(kind='bar',xlabel='Payment Mode',ylabel='Frequency',title='Payment Mode Preferences',color='seagreen')
```

● **Time-based Analysis:**
* Month wise total sales.

```python
order_data['order_date'] = pd.to_datetime(order_data['order_date'])

# Extract month and year from the 'order_date' column
order_data['month_year'] = order_data['order_date'].dt.strftime('%B')
order_data.groupby('month_year')['total_price'].sum().sort_values(ascending=False).plot(kind='bar',xlabel='Month',ylabel='Total Sales',title='Month_wise Total Sales',color='seagreen')
```

* Month and year wise total sales

```python
order_data['order_date'] = pd.to_datetime(order_data['order_date'])
order_data['month_year'] = order_data['order_date'].dt.strftime('%B %Y')
# Grouping by month and year and calculating total sales
monthly_sales = order_data.groupby('month_year')['total_price'].sum().reset_index()
# Extract month and year from the 'order_date' column
plt.figure(figsize=(10, 6))
plt.bar(monthly_sales['month_year'], monthly_sales['total_price'])
plt.title('Year-Wise,Month-Wise Total Sales')
plt.xlabel('Month and Year')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```
